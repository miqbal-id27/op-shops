```python
import json
import time
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

INPUT_FILE = "opshops.json"
OUTPUT_FILE = "opshops_with_coords.json"

def main():
    # Load your existing data
    with open(INPUT_FILE, "r", encoding="utf-8") as f:
        shops = json.load(f)

    geolocator = Nominatim(user_agent="opshop_geocoder")
    geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)  # 1 second delay

    new_shops = []

    for idx, shop in enumerate(shops):
        name = shop[0]
        address = shop[1]

        print(f"[{idx+1}/{len(shops)}] Geocoding:", name)

        lat = None
        lon = None

        try:
            location = geocode(address + ", Australia")
            if location:
                lat = location.latitude
                lon = location.longitude
            else:
                print("  ⚠️ Not found:", address)
        except Exception as e:
            print("  ❌ Error:", e)

        # Insert lat, lon after address (index 2 and 3)
        new_shop = shop.copy()
        new_shop.insert(2, lat)
        new_shop.insert(3, lon)

        new_shops.append(new_shop)

    # Save new JSON
    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(new_shops, f, ensure_ascii=False, indent=2)

    print("✅ Saved to", OUTPUT_FILE)

if __name__ == "__main__":
    main()
```
